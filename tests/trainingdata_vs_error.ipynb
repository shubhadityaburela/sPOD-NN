```python
import numpy as np
import matplotlib.pyplot as plt
from Helper import save_fig
import pandas as pd
import os

impath = "../plots/images_wildfire1D/"
os.makedirs(impath, exist_ok=True) 

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern"]})

SMALL_SIZE = 16   # 16
MEDIUM_SIZE = 18   # 18
BIGGER_SIZE = 20   # 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
beta = np.asarray([5, 10])

err_sPOD_T_1D = np.asarray([0.0378, 0.010])
err_sPOD_S_1D = np.asarray([0.00735, 0.0020])
```

```python
fig, axs = plt.subplots(1, 1, figsize=(8, 5))

axs.plot(beta, err_sPOD_T_1D, color="red", linestyle='--', marker="o", label=r"Temperature")
axs.plot(beta, err_sPOD_S_1D, color="blue", linestyle='--', marker="o", label=r"Supply mass fraction")
# axs.set_yscale('log')
axs.set_title("Number of $\mu$ parameter for training vs error")
axs.grid()
axs.legend(loc='upper right')
axs.set_xticks([5, 10])


fig.supxlabel(r"Number of $\mu$ parameter")
fig.supylabel(r"$E^{\mathrm{sPOD-NN}}_{\mathrm{tot}}$")

save_fig(filepath=impath + 'trainingdata_vs_error_1D', figure=fig)
fig.savefig(impath + "trainingdata_vs_error_1D" + ".eps", format='eps',dpi=600, transparent=True)
```
